```python
from binance import Client
import pandas as pd
import json
```

```python
f = open('config.json')
keys = json.load(f)
```

```python
apikey = keys['apikey']
secret = keys['secret']
client = Client(api_key=apikey, api_secret=secret)
```

```python
tickers = client.get_all_tickers()
tickers
```

```python
historical = client.get_historical_klines('BTCUSDT', Client.KLINE_INTERVAL_1DAY, '1 Jan 2016')
```

```python
hist_df = pd.DataFrame(historical)
```

```python
hist_df
```

```python
hist_df.columns = ['Open Time', 'Open', 'High', 'Low', 'Close', 'Volume', 'Close Time', 'Quote Asset Volume', 
                    'Number of Trades', 'TB Base Volume', 'TB Quote Volume', 'Ignore']
```

```python
hist_df
```

```python
hist_df.dtypes
```

```python
hist_df['Open Time'] = pd.to_datetime(hist_df['Open Time']/1000, unit='s')
hist_df['Close Time'] = pd.to_datetime(hist_df['Close Time']/1000, unit='s')
```

```python
numeric_columns = ['Open', 'High', 'Low', 'Close', 'Volume', 'Quote Asset Volume', 'TB Base Volume', 'TB Quote Volume']
```

```python
hist_df[numeric_columns] = hist_df[numeric_columns].apply(pd.to_numeric, axis=1)
```

```python
hist_df
```

```python
hist_df.set_index('Close Time').tail(100)
```

```python
import mplfinance as mpf
```

```python
mpf.plot(hist_df.set_index('Close Time').tail(120), 
        type='candle', style='charles', 
        volume=True, 
        title='BTCUSDT Last 120 Days', 
        mav=(10,20,30))
```
